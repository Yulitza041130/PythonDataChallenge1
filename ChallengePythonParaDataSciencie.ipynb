### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda1 = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda1.head()
```

#Introducción

El presente análisis tiene como objetivo apoyar al Sr. Juan, propietario de la cadena Alura Store, en la toma de decisión respecto a qué tienda debe vender para iniciar un nuevo emprendimiento. Para ello, se realizó un estudio completo de datos operativos y comerciales de las cuatro tiendas existentes, considerando variables como ingresos, volumen de ventas, calificaciones de clientes, costos de envío y categorías de productos más y menos vendidos.


#1. Análisis de facturación



```python
#Se calcula los ingresos totales por cada tienda

ingreso_tienda1 = tienda1['Precio'].sum()
ingreso_tienda2 = tienda2['Precio'].sum()
ingreso_tienda3 = tienda3['Precio'].sum()
ingreso_tienda4 = tienda4['Precio'].sum()

#Se muestran los resultados por cada tienda

print("Ingreso total tienda 1:", ingreso_tienda1)
print("Ingreso total tienda 2:", ingreso_tienda2)
print("Ingreso total tienda 3:", ingreso_tienda3)
print("Ingreso total tienda 4:", ingreso_tienda4)
```

```python
import matplotlib.pyplot as plt

# Se crea la lista de los ingresos por tienda
ingresos_tiendas = [ingreso_tienda1, ingreso_tienda2, ingreso_tienda3, ingreso_tienda4]
tiendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]

# Creamos el gráfico de columnas (barras verticales)
plt.bar(tiendas, ingresos_tiendas, color='skyblue')  # Definimos el color de las barras

# Añadimos el título y las etiquetas de los ejes
plt.title("Ingresos Totales por Tienda")  # Título del gráfico
plt.xlabel("Tiendas")  # Etiqueta para el eje X
plt.ylabel("Ingreso Total ($)")  # Etiqueta para el eje Y

# Mostrar los valores de ingresos sobre las barras
for i in range(len(ingresos_tiendas)):
    plt.text(i, ingresos_tiendas[i] + 100, f"${ingresos_tiendas[i]:,.2f}", ha="center", fontsize=8.5)

# Mostrar el gráfico
plt.show()
```

La tienda 1 lidera en ingresos totales, superando ligeramente a las demás tiendas. Aunque las diferencias no son extremas, podría ser interesante investigar si esto se debe a mayor volumen de ventas, productos de mayor valor o mejor rendimiento en determinadas categorías. Esta información es clave para orientar estrategias de ventas y evaluar el desempeño individual de cada tienda.

# 2. Ventas por categoría

```python
# Se agrupan los productos vendidos por categoria de cada tienda

categoria_tienda1 = tienda1.groupby('Categoría del Producto')['Precio'].count()
categoria_tienda2 = tienda2.groupby('Categoría del Producto')['Precio'].count()
categoria_tienda3 = tienda3.groupby('Categoría del Producto')['Precio'].count()
categoria_tienda4 = tienda4.groupby('Categoría del Producto')['Precio'].count()

# Se muestran los resultados de ventas por categoría de cada tienda

print("Categorías Tienda 1:\n", categoria_tienda1)
print("\nCategorías Tienda 2:\n", categoria_tienda2)
print("\nCategorías Tienda 3:\n", categoria_tienda3)
print("\nCategorías Tienda 4:\n", categoria_tienda4)

```

```python
# Se combinan los datos de todas las tiendas
categoria_total = categoria_tienda1 + categoria_tienda2 + categoria_tienda3 + categoria_tienda4

# Se ordenan las categorías de mayor a menor
categoria_total = categoria_total.sort_values(ascending=False)

# Gráfica de barras
plt.figure(figsize=(10, 6))
categoria_total.head(10).plot(kind='bar', color='skyblue')

# Etiquetas y título
plt.title('Top 10 Categorías Más Vendidas en Todas las Tiendas', fontsize=16)
plt.xlabel('Categoría', fontsize=12)
plt.ylabel('Cantidad de Cuotas Vendidas', fontsize=12)
plt.xticks(rotation=45, ha='right')

# Mostramos el gráfico
plt.tight_layout()
plt.show()
```



Las categorías con mayor volumen de ventas en general son **Muebles** y **Electrónicos**, mostrando una demanda alta y constante en las cuatro tiendas. En particular:

- **Muebles** lidera en todas las tiendas, especialmente en la Tienda 3 (499 ventas) y Tienda 4 (480 ventas).
- **electrónicos** mantiene cifras elevadas y muy parejas entre tiendas, alcanzando el mismo valor (451) en Tienda 3 y Tienda 4.
- Las categorías con menor volumen son **instrumentos musicales**, **libros** y **articulos para el hogar**, lo que puede indicar un menor interés por estos productos o una menor oferta.

Este patrón sugiere que los consumidores tienen una clara preferencia por productos grandes y tecnológicos. Es importante considerar estas tendencias para tomar cualquier tipo de decisión.


# 3. Calificación promedio de la tienda


```python
# Se obtiene la calificación promedio para cada tienda

calificacion_prom_tienda1 = round(tienda1["Calificación"].mean(), 2)
calificacion_prom_tienda2 = round(tienda2["Calificación"].mean(), 2)
calificacion_prom_tienda3 = round(tienda3["Calificación"].mean(), 2)
calificacion_prom_tienda4 = round(tienda4["Calificación"].mean(), 2)


# Se muestran las calificaciones promedio de cada tienda

print("Calificación promedio tienda 1:", calificacion_prom_tienda1)
print("\nCalificación promedio tienda 2:", calificacion_prom_tienda2)
print("\nCalificación promedio tienda 3:", calificacion_prom_tienda3)
print("\nCalificación promedio tienda 4:", calificacion_prom_tienda4)
```

```python
# Calificaciones promedio
calificaciones = [calificacion_prom_tienda1, calificacion_prom_tienda2, calificacion_prom_tienda3, calificacion_prom_tienda4]
tiendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]

# Gráfico de barras (columnas)
plt.bar(tiendas, calificaciones, color="lightblue")

# Gráfico de líneas sobre el gráfico de barras
plt.plot(tiendas, calificaciones, color="green", marker=".", linestyle='-', linewidth=1)

# Agregamos título y etiquetas
plt.title("Calificación Promedio por Tienda")
plt.ylabel("Calificación")

# Mostramos los valores sobre cada barra
for i in range(len(tiendas)):
    plt.text(i, calificaciones[i] + 0.02, str(calificaciones[i]), ha="center")

# Mostrar gráfico
plt.show()

```



Las calificaciones promedio de las tiendas oscilan entre **3.98 y 4.05**, lo que indica un nivel de satisfacción del cliente relativamente alto y homogéneo entre las cuatro tiendas. La **Tienda 3** presenta la calificación más alta (**4.05**), seguida muy de cerca por la **Tienda 2** (**4.04**) y la **Tienda 4** (**4.00**). Por su parte, la **Tienda 1** tiene la calificación más baja (**3.98**), aunque la diferencia con el resto no es significativa.

Esto sugiere que **la experiencia del cliente es consistente**, pero pequeñas diferencias podrían indicar áreas de mejora específicas en la Tienda 1, o destacar buenas prácticas que podrían ser replicadas desde la Tienda 3.
